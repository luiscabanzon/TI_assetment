## Second Exercise: Top 10 flight arrivals in 2013
Now it's time to get deeper into data. These are the steps to follow:

* Filter the data by year (2013). Panda is really powerful working with dates. 
* Group flights by arrivals.
* Sort by number of passengers and return the top10 list.

But before this, we need to have a better understanding of the file database structure.
We can start by open a sample of this file with all the columns and some rows.

```python
bookings_sample = pd.io.parsers.read_csv("D:/data/bookings.csv.bz2", compression = "bz2", error_bad_lines=False, sep = "^", nrows = 5)
list(bookings_sample.columns.values)
```

```python
len(list(bookings_sample.columns.values))
```

```python
bookings_sample.head()
```

From this quick view, we can identify the 3 columns that are relevant to us right now:
* arr_port : Destination airport
* brd_time : Flight board date and hour.
* pax : Number of people included in that booking (negative numbers mean cancellations)


```python
bookings = pd.io.parsers.read_csv("D:/data/bookings.csv.bz2", compression = "bz2", error_bad_lines= False, sep = "^", usecols = ("arr_port","brd_time           ","pax"), index_col = 1, parse_dates = [True])
```

```python
bookings.head()
```


Now, let's group passengers by destination and sort these airports by traffic:


```python
arrivals_per_port = bookings[["arr_port","pax"]].groupby(["arr_port"]).sum()
arrivals_per_port.sort(columns = "pax" , ascending = False).head(10)
```
