Now it's time to get deeper into data. These are the steps to follow:

From this quick view, we can identify the 3 columns that are relevant to us right now:
* arr_port : Destination airport
* brd_time : Flight board date and hour.
* pax : Number of people included in that booking (negative numbers mean cancellations)


```python
bookings = pd.io.parsers.read_csv("D:/data/bookings.csv.bz2", compression = "bz2", error_bad_lines= False, sep = "^", usecols = ("arr_port","brd_time           ","pax"), index_col = 1, parse_dates = True)
```

```python
bookings.head()
```


Now, let's group passengers by destination and sort these airports by traffic:


```python
arrivals_per_port = bookings[["arr_port","pax"]].groupby(["arr_port"]).sum()
arrivals_per_port.sort(columns = "pax" , ascending = False).head(10)
```

  
Good, but not all of us know IATA codes by heart, right?
No worries... **GeoBases** reporting for duty!
A couple of comments:
* Some IATA codes were no clean, (additional spaces), and to solve this we use .string().
* In some cases the code provided was not from an airport but from a country, that's why I load 2 databases ("ori_port" and "countries".


```python
from GeoBases import GeoBase
geo_o = GeoBase(data='ori_por', verbose=False)
geo_c = GeoBase(data='countries', verbose=False)

arrivals_per_port["arr_name"] = ""
                  
for i in range(len(arrivals_per_port)):
    if  arrivals_per_port.index[i].strip() in geo_o:
        arrivals_per_port.ix[i,"arr_name"] =  geo_o.get(arrivals_per_port.index[i].strip(), "name")
    else:
        arrivals_per_port.ix[i,"arr_name"] =  geo_c.get(arrivals_per_port.index[i].strip(), "name")
        
arrivals_per_port.sort(columns = "pax" , ascending = False).head(10)
```

## Third Exercise: Monthly Searches (Malaga, Madrid, Barcelona)
---
Time to search into Searches! First of all, we need to get familiar with the file.

```python
searches_sample = pd.io.parsers.read_csv("D:/data/searches.csv.bz2", compression = "bz2", error_bad_lines=False, sep = "^", nrows = 5)
searches_sample
```

```python
searches = pd.io.parsers.read_csv("D:/data/searches.csv.bz2", compression = "bz2", error_bad_lines=False, sep = "^", usecols = [0,6])
```

```python
searches.head()
```

Now the steps to follow are:
* Extract entries per destination (Madrid : "MAD", Barcelona :"BCN" , Malaga: "AGP").
* Translate dates into pandas.datetime for(for some reason, the arguments with this purpose in read_csv don't work properly.
* Join the monthly consolidated data and plot searches per destination.

```python
searches_MAD = searches[searches["Destination"] == "MAD"]
searches_BCN = searches[searches["Destination"] == "BCN"]
searches_AGP = searches[searches["Destination"] == "AGP"]
```

```python
searches_BCN["Date"] = searches_BCN["Date"].apply(lambda x: pd.to_datetime(x, infer_datetime_format = True, coerce= True))
```

```python
searches_AGP["Date"] = searches_AGP["Date"].apply(lambda x: pd.to_datetime(x, infer_datetime_format = True, coerce= True))
```

```python
searches_MAD["Date"] = searches_MAD["Date"].apply(lambda x: pd.to_datetime(x, infer_datetime_format = True, coerce= True))
```

```python
searches_MAD.index = searches_MAD["Date"]
searches_BCN.index = searches_BCN["Date"]
searches_AGP.index = searches_AGP["Date"]
```

```python
searches_MAD.__delitem__("Date")
searches_BCN.__delitem__("Date")
searches_AGP.__delitem__("Date")
```

```python
MAD = searches_MAD.groupby(searches_MAD.index.month).count()
BCN = searches_BCN.groupby(searches_BCN.index.month).count()
AGP = searches_AGP.groupby(searches_AGP.index.month).count()
```

```python
MAD.columns = ["MAD"] ; MAD["month"] = MAD.index
BCN.columns = ["BCN"] ; BCN["month"] = BCN.index
AGP.columns = ["AGP"] ; AGP["month"] = AGP.index
```

```python
searches_month = pd.merge(MAD,BCN)
```

```python
searches_month = pd.merge(searches_month, AGP)
```

```python
searches_month.index = searches_month["month"]
searches_month.__delitem__("month")
```

```python
searches_month
```

```python
import matplotlib.pyplot as plt
searches_month.plot()
```

Some conclusions from this graph:
 * The flights to Malaga are not as numerous compared with Madrid and Barcelona, but they are more stable throughout the year.
 * The number of flights to Madrid and Barcelona follow the same trend and at the end of the year they become much closer, with a significant decrease.
